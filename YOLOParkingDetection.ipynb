# **YOLOParkingDetection: Fine-Tuned YOLOv8 for Parking Spot Detection**

## **Project Overview**

**YOLOParkingDetection** is a deep learning project designed to detect **parking spot availability** in images, classifying each spot as either **Empty** or **Occupied**.  
The workflow included **data exploration & visualization**, dataset preparation, **fine-tuning YOLOv8**, and deployment as an interactive **Streamlit application** and publishing on **Hugging Face**.

## **Objective**

Develop and deploy a robust **object detection model** to support **smart parking management systems**, enabling real-time monitoring of **available and occupied parking spaces** in urban environments.

## **Modeling Approach**

- **Base Model**: **YOLOv8** (Ultralytics implementation, pretrained on COCO dataset)  
- **Fine-Tuning**: Adapted on the parking dataset with transfer learning  
- **Prediction Outputs**: Bounding boxes around detected parking spots labeled as **Empty** or **Occupied**

## **Performance**

The fine-tuned model achieved outstanding results on the test set:  
- **Precision**: `99.39%`  
- **Recall**: `98.20%`  
- **mAP@50**: `98.73%`  
- **mAP@50-95**: `97.46%`  
- **Speed**: ~10.7ms per image (real-time capable)

#### **Duplicate Image Detection**

```python
def file_hash(path):
    """Return MD5 hash of a file."""
    with open(path, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()

def check_duplicates(dataset_dir, split):
    """Check for duplicate images within a split."""
    images = glob.glob(f"/kaggle/working/{dataset_dir}/{split}/images/*.jpg")
    hashes, dupes = {}, []
    
    for img in images:
        h = file_hash(img)
        if h in hashes:
            dupes.append((img, hashes[h]))
        else:
            hashes[h] = img

    if dupes:
        print(f"⚠️ Duplicates found in {split}: {len(dupes)}")
        for d in dupes:
            print(" -", d)
    else:
        print(f"✅ No duplicates in {split} ({len(images)} images)")

for split in ["train", "valid", "test"]:
    check_duplicates("Parking-Detection-1", split)
```

#### **Check for Corrupted Files**

```python
def check_corrupted(dataset_dir, split):
    """Check for corrupted images in a split."""
    images = glob.glob(f"/kaggle/working/{dataset_dir}/{split}/images/*.jpg")
    corrupted = []
    
    for img_path in images:
        try:
            with Image.open(img_path) as img:
                img.verify()  # raises error if corrupted
        except Exception:
            corrupted.append(img_path)
    
    if corrupted:
        print(f"⚠️ Corrupted images found in {split}: {len(corrupted)}")
        for c in corrupted:
            print(" -", c)
    else:
        print(f"✅ No corrupted images in {split} ({len(images)} images)")

for split in ["train", "valid", "test"]:
    check_corrupted("Parking-Detection-1", split)
```

#### **Visualize Sample Images**

```python
class_names = ["empty", "occupied"]
class_colors = {"empty": (0, 255, 0), "occupied": (255, 0, 0)} 

def plot_random_annotated_samples(dataset_dir, split="train", n_samples=6, per_row=3):
    """Show random annotated samples from YOLO dataset split."""
    image_dir = os.path.join(f"/kaggle/working/{dataset_dir}", split, "images")
    label_dir = os.path.join(f"/kaggle/working/{dataset_dir}", split, "labels")
    images = os.listdir(image_dir)
    
    random.seed() 
    sample_imgs = random.sample(images, min(n_samples, len(images)))
    
    rows = (len(sample_imgs) + per_row - 1) // per_row
    fig, axes = plt.subplots(rows, per_row, figsize=(15, 5 * rows))
    axes = axes.flatten()
    
    for i, img_file in enumerate(sample_imgs):
        img_path = os.path.join(image_dir, img_file)
        label_path = os.path.join(label_dir, img_file.replace(".jpg", ".txt"))
        
        img = cv2.imread(img_path)
        h, w, _ = img.shape
        
        if os.path.exists(label_path):
            with open(label_path) as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) >= 5:
                        cls, x, y, bw, bh = map(float, parts[:5])
                        cls = int(cls)
                        x, y, bw, bh = x*w, y*h, bw*w, bh*h
                        x1, y1 = int(x - bw/2), int(y - bh/2)
                        x2, y2 = int(x + bw/2), int(y + bh/2)

                        color = class_colors[class_names[cls]]
                        cv2.rectangle(img, (x1,y1), (x2,y2), color, 2)
                        cv2.putText(img, class_names[cls], (x1,y1-5),
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
        
        axes[i].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axes[i].axis("off")
    
    # Hide unused axes properly
    for j in range(len(sample_imgs), len(axes)):
        axes[j].axis("off")
    
    plt.suptitle(f"Random annotated samples from {split} set", fontsize=16, fontweight="bold")
    plt.show()

# Example usage
plot_random_annotated_samples("Parking-Detection-1", split="train", n_samples=6, per_row=3)
```

### **Training YOLOv8 for Parking Spot Detection**

#### **About YOLOv8**
**YOLOv8 (You Only Look Once, version 8)** is a state-of-the-art **object detection architecture** developed by Ultralytics.  
It is designed for **real-time detection** with high accuracy and efficiency, making it suitable for applications like traffic monitoring, surveillance, and parking management.  

YOLOv8 processes the entire image in a **single forward pass**, predicting **bounding boxes** and **class labels** simultaneously.  
This enables it to detect whether a parking spot is **Empty** or **Occupied** in just milliseconds per image.  

#### **Why Fine-Tuning with YOLOv8?**
Training YOLOv8 from scratch requires massive datasets and computational resources.  
Instead, this project leverages **transfer learning**, fine-tuning YOLOv8 on a **custom parking lot dataset** to specialize in detecting parking spot availability.  

This approach:  
- Provides **real-time detection** of parking occupancy  
- Enhances **accuracy** even with a limited dataset  
- Improves **generalization** to new parking lot environments  
- Reduces **training time** while maintaining high performance  

In this project, YOLOv8 serves as a powerful **object detector**, fine-tuned to classify parking spaces as:  
- **Empty**  
- **Occupied**  

This enables the development of **smart parking systems** that can assist cities, malls, and facilities in **optimizing parking space usage** and improving **urban mobility**.


```python
model = YOLO("yolov8n.pt")

model.train(
    data="/kaggle/working/Parking-Detection-1/data.yaml",
    epochs=50,
    imgsz=640 
)
```

#### **Model Evaluation**

```python
print("🔹 Validation Set Metrics")
val_metrics = model.val(split="val") 
print("Precision:", val_metrics.box.p)
print("Recall:", val_metrics.box.r)
print("mAP50:", val_metrics.box.map50)
print("mAP50-95:", val_metrics.box.map)
```

```python
print("\n🔹 Test Set Metrics")
test_metrics = model.val(split="test")  # test split from data.yaml
print("Precision:", test_metrics.box.p)
print("Recall:", test_metrics.box.r)
print("mAP50:", test_metrics.box.map50)
print("mAP50-95:", test_metrics.box.map)
```

```python
val_scores = {
    "Precision": val_metrics.box.p.mean(),
    "Recall": val_metrics.box.r.mean(),
    "mAP50": val_metrics.box.map50.mean(),
    "mAP50-95": val_metrics.box.map.mean()
}

test_scores = {
    "Precision": test_metrics.box.p.mean(),
    "Recall": test_metrics.box.r.mean(),
    "mAP50": test_metrics.box.map50.mean(),
    "mAP50-95": test_metrics.box.map.mean()
}


labels = list(val_scores.keys())
val_values = list(val_scores.values())
test_values = list(test_scores.values())

x = range(len(labels))
width = 0.35

plt.figure(figsize=(8,6))
plt.bar([i - width/2 for i in x], val_values, width, label="Validation", color="skyblue")
plt.bar([i + width/2 for i in x], test_values, width, label="Test", color="salmon")

plt.xticks(x, labels, fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.ylim(0, 1.1)
plt.title("Validation vs Test Performance", fontsize=14, fontweight="bold")
plt.legend()
plt.show()
```

```python
cm_val = val_metrics.confusion_matrix.matrix[:2, :2]
cm_test = test_metrics.confusion_matrix.matrix[:2, :2]

class_names = ["empty", "occupied"]

plt.figure(figsize=(5,4))
sns.heatmap(cm_val.astype(int), annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.title("Validation Confusion Matrix")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(cm_test.astype(int), annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names, yticklabels=class_names)
plt.title("Test Confusion Matrix")
plt.show()
```

```python
test_dir = "Parking-Detection-1/test/images"

random.seed() 
sample_imgs = random.sample(
    [os.path.join(test_dir, f) for f in os.listdir(test_dir) if f.endswith((".jpg",".png"))],
    1
)

for img_path in sample_imgs:
    results = model.predict(img_path, conf=0.25, imgsz=640)
    
    results[0].show()
    
    pred_classes = results[0].boxes.cls.cpu().numpy().astype(int)
    
    names = model.names  
    labels = [names[c] for c in pred_classes]
    
    empty_count = labels.count("empty")
    occupied_count = labels.count("occupied")
    
    print(f"Image: {os.path.basename(img_path)}")
    print(f"Empty: {empty_count}, Occupied: {occupied_count}")
```

```python
def detect_parking(model, img_path, conf=0.25, imgsz=640):
    """
    Detect parking spots in an image using YOLO model.
    Displays the image with bounding boxes and prints counts of empty and occupied spots.
    """
    results = model.predict(img_path, conf=conf, imgsz=imgsz, verbose=False)
    
    img_bgr = results[0].plot()   
    img_rgb = img_bgr[:, :, ::-1] 
    
    plt.figure(figsize=(8, 8))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(f"Model Prediction")
    plt.show()
    
    pred_classes = results[0].boxes.cls.cpu().numpy().astype(int)
    names = model.names
    labels = [names[c] for c in pred_classes]
    
    empty_count = labels.count("empty")
    occupied_count = labels.count("occupied")
    
    print(f"🅿️ Empty spots: {empty_count}")
    print(f"🚗 Occupied spots: {occupied_count}")
    
    return empty_count, occupied_count
```
